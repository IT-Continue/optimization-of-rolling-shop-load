```python
import numpy as np
def get_best_assignment(assignments, assignments_time_ovens, series_list):
    if len(assignments)==0:
        return [], assignments_time_ovens
    min_val = np.inf
    res = -1
    for i in range(len(assignments)):
        if sum(assignments_time_ovens[i]) < min_val:
            res = i
            min_val = sum(assignments_time_ovens[i])
    return assignments[res], assignments_time_ovens[res], series_list[res]

get_best_assignment([{1}, {2}], [[1, 2, 3], [1, 1, 1]], [1, 2])
```
